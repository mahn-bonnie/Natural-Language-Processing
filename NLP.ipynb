```python
import pyglet
from pyglet.gl import *

# Load texture
texture = pyglet.image.load("terrain_texture.jpg").get_texture()

# Declare variables
torus_radius = 100
torus_thickness = 30
texture_scale = 5.0
texture_rotation = 0.0

# Initialize window
window = pyglet.window.Window(800, 600, resizable=True)
glEnable(GL_DEPTH_TEST)
glEnable(GL_TEXTURE_2D)

# Set up camera
glMatrixMode(GL_PROJECTION)
glLoadIdentity()
gluPerspective(45, window.width / window.height, 0.1, 1000)
glMatrixMode(GL_MODELVIEW)
glLoadIdentity()
glTranslatef(0, 0, -500)

# Function to draw the torus with texture
def draw_torus_with_texture():
    glBindTexture(texture.target, texture.id)
    glBegin(GL_QUADS)

    for theta in range(0, 360, 10):
        for phi in range(0, 360, 10):
            u = theta / 360.0
            v = phi / 360.0
            x = (torus_radius + torus_thickness * cos(radians(theta))) * cos(radians(phi))
            y = (torus_radius + torus_thickness * cos(radians(theta))) * sin(radians(phi))
            z = torus_thickness * sin(radians(theta))
            glTexCoord2f(u * texture_scale, v * texture_scale)
            glVertex3f(x, y, z)

    glEnd()

# Event handlers
@window.event
def on_draw():
    window.clear()
    glPushMatrix()
    glRotatef(texture_rotation, 0, 0, 1)
    draw_torus_with_texture()
    glPopMatrix()

@window.event
def on_resize(width, height):
    glViewport(0, 0, width, height)
    glMatrixMode(GL_PROJECTION)
    glLoadIdentity()
    gluPerspective(45, width / height, 0.1, 1000)
    glMatrixMode(GL_MODELVIEW)
    return pyglet.event.EVENT_HANDLED

# pyglet.app.run() starts the application loop
pyglet.app.run()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def initialize_weights(n, non_zero_diagonal=False, asymmetry=False):
    """Initializes the weight matrix for a Hopfield network.

    Args:
        n (int): Number of units in the network.
        non_zero_diagonal (bool, optional): Whether to include non-zero diagonal elements. Defaults to False.
        asymmetry (bool, optional): Whether to introduce asymmetry in the weights. Defaults to False.

    Returns:
        numpy.ndarray: The weight matrix with the specified properties.
    """

    weights = np.random.rand(n, n) * 2 - 1  # Random values between -1 and 1
    weights = (weights + weights.T) / 2  # Make symmetric
    if not non_zero_diagonal:
        np.fill_diagonal(weights, 0)  # Set diagonal elements to 0
    if asymmetry:
        # Introduce asymmetry (can be adjusted for specific effects)
        for i in range(n):
            weights[i, (i + 1) % n] += 0.2
            weights[(i - 1) % n, i] -= 0.2
    return weights

def update_unit(i, weights, states):
    """Updates the state of a unit based on the weighted inputs from other units.

    Args:
        i (int): Index of the unit to update.
        weights (numpy.ndarray): The weight matrix of the network.
        states (numpy.ndarray): The current states of all units.

    Returns:
        int: The updated state of the unit (1 or -1).
    """

    net_input = np.dot(weights[i], states) - states[i] * weights[i, i]  # Consider diagonal if allowed
    return 1 if net_input > 0 else -1

def update_network(weights, states, max_iterations=100):
    """Updates the states of all units in the network iteratively until convergence.

    Args:
        weights (numpy.ndarray): The weight matrix of the network.
        states (numpy.ndarray): The initial states of all units.
        max_iterations (int, optional): Maximum number of iterations. Defaults to 100.

    Returns:
        numpy.ndarray: The final states of all units after convergence.
    """

    converged = False
    iterations = 0
    while not converged and iterations < max_iterations:
        new_states = np.copy(states)
        for i in range(len(states)):
            new_states[i] = update_unit(i, weights, states)
        if (new_states == states).all():
            converged = True
        states = new_states
        iterations += 1
    return states

def find_stable_states(weights):
    """Finds all stable states (attractors) of the Hopfield network.

    Args:
        weights (numpy.ndarray): The weight matrix of the network.

    Returns:
        list: A list of numpy.ndarray, where each element represents a stable state.
    """

    visited_states = set()
    stable_states = []

    def explore_state(state):
        if state in visited_states:
            return
        visited_states.add(state)
        next_state = update_network(weights, state)
        if np.array_equal(state, next_state):
            stable_states.append(state)
            return
        else:
            explore_state(next_state)

    for i in range(2**len(weights)):
        initial_state = np.array([1 if (i >> j) & 1 else -1 for j in range(len(weights))])
        explore_state(initial_state)

    return stable_states

def analyze_results(n, num_networks, non_zero_diagonal=False, asymmetry=False):
    """Analyzes the stable states and unit activation frequencies for multiple networks.

    Args:
        n (int): Number of units in the network.
        num_networks (int): Number of networks to analyze.
        non_zero_diagonal (bool, optional): Whether non-zero diagonal elements are used. Defaults to False.
        asymmetry (bool, optional): Whether asymmetry is introduced in the weights."""
def main():
    """Main function to run the Hopfield network analysis."""

    n = 3  # Number of units in the network
    num_networks = 100  # Number of networks to analyze
    iterations = 1000  # Maximum iterations per network

    # Analyze networks with different weight configurations
    configurations = [
        ("Random, Symmetric", False, False),
        ("Non-zero Diagonal", True, False),
        ("Asymmetric", False, True),
    ]

    for label, non_zero_diagonal, asymmetry in configurations:
        print(f"\nAnalysis for: {label}")

        # Analyze multiple networks per configuration
        stable_states_counts = np.zeros((n + 1,), dtype=int)
        unit_activation_freqs = np.zeros((n, 2), dtype=int)  # Track on/off counts for each unit

        for _ in range(num_networks):
            weights = initialize_weights(n, non_zero_diagonal, asymmetry)

            # Find stable states and update counts
            for stable_state in find_stable_states(weights):
                stable_states_counts[np.sum(stable_state == 1)] += 1
                unit_activation_freqs += np.outer(stable_state, stable_state + 1)

            # Optional visualization of network dynamics and stable states
            # visualize_network_dynamics(weights, states)
            # visualize_stable_states(weights, stable_states)

        # Normalize unit activation frequencies
        unit_activation_freqs /= stable_states_counts.sum()

        # Print and plot results
        print("Stable states frequencies:")
        print(stable_states_counts)
        plt.figure()
        plt.bar(range(n + 1), stable_states_counts)
        plt.xlabel("Number of stable states")
        plt.ylabel("Frequency")
        plt.title(f"Stable States Frequency Distribution ({label})")
        plt.show()

        print("Unit activation frequencies:")
        print(unit_activation_freqs)
        plt.figure()
        for i in range(n):
            plt.bar([0, 1], unit_activation_freqs[i])
            plt.xlabel("Unit state (0/1)")
            plt.ylabel("Frequency")
            plt.title(f"Unit {i+1} Activation Frequency ({label})")
            plt.show()

if __name__ == "__main__":
    main()


```

```python
import numpy as np

def hopfield_network_s(n, s):
    # Initialize weights matrix with symmetric weights around the diagonal
    weights = np.random.uniform(-1, 1, size=(n, n))
    weights = 0.5 * (weights + weights.T)  # Make the matrix symmetric

    # Relax constraint S by adding random deviations to corresponding weights
    weights += np.random.uniform(-s, s, size=(n, n))

    return weights

# Example Usage:
n = 5  # Size of the Hopfield network
s_values = [0.1, 0.5, 0.9]  # Different levels of relaxation

for s in s_values:
    weights = hopfield_network_s(n, s)
    print(f"Weights for relaxation S={s}:\n{weights}\n")
```

```python
# Hopfield Network Simulation with Relaxed Constraint Z

import numpy as np
import matplotlib.pyplot as plt

def hopfield_network(input_patterns, relaxation_factor):
    n_units = input_patterns.shape[1]

    # Initialize weights with random values
    weights = np.random.rand(n_units, n_units)

    # Relax constraint Z by adjusting diagonal weights
    weights[np.diag_indices(n_units)] = np.random.uniform(-relaxation_factor, relaxation_factor, n_units)

    # Update weights based on input patterns
    for pattern in input_patterns:
        weights += np.outer(pattern, pattern)

    # Apply signum function to determine states
    states = np.sign(np.dot(input_patterns, weights))

    return states

# Experiment Parameters
input_patterns = np.array([[1, -1, 1], [-1, 1, -1], [1, 1, 1]])  # Example input patterns
relaxation_factors = [0.1, 0.5, 0.9]  # Different relaxation levels

# Run the experiment
for factor in relaxation_factors:
    result_states = hopfield_network(input_patterns, factor)
    print(f"Relaxation Factor {factor}: {result_states}")
```

```python
# Hopfield Network Simulation with Relaxed Constraint S

import numpy as np
import matplotlib.pyplot as plt

def hopfield_network(input_patterns, symmetry_factor):
    n_units = input_patterns.shape[1]

    # Initialize weights with random values
    weights = np.random.rand(n_units, n_units)

    # Relax constraint S by adjusting weights asymmetrically
    weights = 0.5 * (weights + weights.T)  # Ensure symmetry
    weights += np.random.uniform(-symmetry_factor, symmetry_factor, (n_units, n_units))

    # Update weights based on input patterns
    for pattern in input_patterns:
        weights += np.outer(pattern, pattern)

    # Apply signum function to determine states
    states = np.sign(np.dot(input_patterns, weights))

    return states

# Experiment Parameters
input_patterns = np.array([[1, -1, 1], [-1, 1, -1], [1, 1, 1]])  # Example input patterns
symmetry_factors = [0.1, 0.5, 0.9]  # Different symmetry relaxation levels

# Run the experiment
for factor in symmetry_factors:
    result_states = hopfield_network(input_patterns, factor)
    print(f"Symmetry Factor {factor}: {result_states}")
```
